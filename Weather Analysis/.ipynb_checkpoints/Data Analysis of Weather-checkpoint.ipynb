```python
# describe all instances where 'Weather is Clear' or 'visibilty is above 40'
df[(df.Weather == 'Clear') | (df['Visibility_km'] > 40)]
```

```python
# find all instances where 'Weather is Clear' and 'humidity is greater than 50' or 'visibility is above 40'
df[
((df.Weather == 'Clear') & (df['Rel Hum_%'] > 50)) | (df['Visibility_km'] > 40)
]
```
